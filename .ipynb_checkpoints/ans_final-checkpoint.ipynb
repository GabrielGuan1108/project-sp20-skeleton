The MST pruning algorithm:

```python
# whole code for MST pruning algorithm

# Given a Tree and original graph G, remove the node which will decrease the pair wise cost of T most.
# If no such node exists, return T
# It's a slow algorithm, O(n^3) ?
def pruning(G,T):
        while True:
            # Special cases handling
            if len(T) == 1 and is_valid_network(G,T):
                return T
            if len(T) == 1 and not is_valid_network(G,T):
                print('error: T has size 1 and not valid')
                return(T)
            
            # Get all leaves of T, O(n)
            cost = average_pairwise_distance(T)
            leaves = [x for x in T.nodes() if T.degree(x)==1]
            index = float('inf')
            
            # Iterate through all leaves to find min T
            for leaf in leaves:
                Tnew = T.copy()
                Tnew.remove_node(leaf)
                try:
                    valid, c = is_valid_network(G,Tnew), average_pairwise_distance(Tnew)
                except ZeroDivisionError:
                    print('zero division: ', path)
                    continue
                
                if len(Tnew) == 1 and valid:
                    return T
                
                if valid and c < cost:
                    cost, index = c, leaf
                    
            # If no leaves are removed during the iteration, return T
            if index == float('inf') and is_valid_network(G,T):
                return T
    
            if not index == float('inf'):
                T.remove_node(index)
                
def MSTalgorithm(G):
    # Get the MST of G by networkx minimum_spanning_edges method
    G = G.copy()
    T = nx.Graph()
    T.add_nodes_from(G.nodes(data=True))
    T.add_edges_from(nx.minimum_spanning_edges(G))
    
    # Remove useless nodes
    return pruning(G, T)
```

When we are reading our output and comparing them with other team's score, we find that there are a lot of inputs has a single node connected with all other nodes. So before running MST pruning algorithm, we need to check if such a node exists. If exists, just return a single tree with only that node will give us a minimium cost of zero.

```python
# check if we can solve this by using a single vertex
# O(n) runtime
def checkZero(G):
    exists = [x for x in G.nodes() if G.degree(x)==(len(G.nodes())-1)]
    if len(exists) != 0:
        return (True, exists[0])
    return (False, 0)
```

An algorithm, find nodes with largest degree and create its shorest path tree(O(n)), and then remove nodes to minimize cost(O(n^3)).

```python
# Create a spt given Graph and a node
# Return: a spt rooted on noded of G
def create_spt(G,node):
    G = G.copy()
    T = nx.Graph()
    T.add_nodes_from(G.nodes(data=True))
    length, pathes = nx.single_source_dijkstra(G, node)
    for path in pathes:
        for i in range(0, len(pathes[path])-1):
            T.add_edge(pathes[path][i],  pathes[path][i+1])
    return T
```

```python
%%time
def spt(G):
    # check if a node connected with all other nodes exists
    complete, node = checkZero(G)
    if complete:
        T = nx.Graph()
        T.add_node(node)
        return T
    # find nodes with max degree
    node = []
    for i in range(len(G)-1, -1, -1):
        node = [x for x in G.nodes() if G.degree(x)==i]
        if node:
            break

            
    # Create the shorest path Tree root on max degree nodes by Dijkstra
    best, cost = None, float('inf')
    for root in node:
        T = create_spt(G,root)

        # Remove all the leaf nodes first (can remove a lot useless nodes because our SPT is shallow)
        T.remove_nodes_from([x for x in T.nodes() if T.degree(x)==1])
        #print('removed all leaves')


        # remove useless nodes from SPT to minimize cost
        #print('enter pruning') 
        T = pruning(G, T)
        Tcost = average_pairwise_distance(T)
        if Tcost < cost:
            cost = Tcost
            best = T   
        
    return best
    
    
# Same as spt, but it doesn't iterate through all nodes with max degree. It pick up a random one. 
# Faster than spt.
def spt_fast(G):
    for i in range(len(G)-1, -1, -1):
        node = [x for x in G.nodes() if G.degree(x)==i]
        if node:
            break

            
    best, cost = None, float('inf')
    T = create_spt(G,node[0])

    T.remove_nodes_from([x for x in T.nodes() if T.degree(x)==1])
    return pruning(G, T)

path = './inputs/large-104.in'
G = read_input_file(path,float('inf'))

spt(G)
```

Here is the code for the final solve method. (it is the combination of MST pruning and Zero check, I will call it Final Solve below)

```python
#final solve method
def solve(G):
    """
    Args:
        G: networkx.Graph

    Returns:
        T: networkx.Graph
    """

    # TODO: your code here!
    zero, node = checkZero(G)
    if zero:
        T = nx.Graph()
        T.add_nodes_from([node])
        return T
        
    T_MST, T_SPT = MSTalgorithm(G), spt(G)
    if average_pairwise_distance(T_MST) < average_pairwise_distance(T_SPT):
        return T_MST
    else: 
        return T_SPT
```
