So before running MST purning algorithm, we need to check if such a node exists. If exists, just return a single tree with only that node will give us a minimium cost of zero.

```python
#check if we can solve this by using a single vertex
def checkZero(G):
    exists = [x for x in G.nodes() if G.degree(x)==(len(G.nodes())-1)]
    print(exists)
    if len(exists) != 0:
        return (True, exists[0])
    return (False, 0)

path = './inputs/large-102.in'
G = read_input_file(path,float('inf'))

print(checkZero(G))

```

Here is the code for the final solve method. (it is the combination of MST purning and Zero check, I will call it Final Solve below)

```python
#final solve method
def solve(G):
    """
    Args:
        G: networkx.Graph

    Returns:
        T: networkx.Graph
    """

    # TODO: your code here!
    zero, node = checkZero(G)
    if zero:
        T = nx.Graph()
        T.add_nodes_from([node])
        return T
        
    T = MSTalgorithm(G)
    if not is_valid_network(G,T):
        print('output not valid')
    return T
```

```python
#这个cell是我朋友的码 跑的快一点50名左右
def alg2(G):
    """
    Args:
        G: networkx.Graph

    Returns:
        T: networkx.Graph
    """

    # TODO: your code here!
    # check complete graph
    m = G.number_of_edges()
    n = G.number_of_nodes()
    if m >= n * (n - 1) / 2 - 1:
        T = nx.Graph()
        T.add_node(0)
        return T
    # Minimum Spanning Tree T, our output
    T = nx.Graph()
    T.add_nodes_from(G.nodes(data = True))
    T.add_edges_from(nx.minimum_spanning_edges(G))
    # All the leave nodes
    leaves = [x for x in T.nodes() if T.degree(x) == 1]
    #leaves.sort(reverse=True)
    # Delete leaves from MST
    for i in leaves:
        temp = T.copy()
        temp.remove_node(i)
        if average_pairwise_distance(temp) < average_pairwise_distance(T):
            T = temp
    # another 2 rounds of deleting leaves from T
    leaves = [x for x in T.nodes() if T.degree(x) == 1]
    #leaves.sort(reverse=True)
    for i in leaves:
        temp = T.copy()
        temp.remove_node(i)
        if is_valid_network(G, temp) and average_pairwise_distance(temp) < average_pairwise_distance(T):
            T = temp
    leaves = [x for x in T.nodes() if T.degree(x) == 1]
    #leaves.sort(reverse=True)
    for i in leaves:
        temp = T.copy()
        temp.remove_node(i)
        if is_valid_network(G, temp) and average_pairwise_distance(temp) < average_pairwise_distance(T):
            T = temp
    return T
```
