```python
import networkx as nx
from parse import read_input_file, write_output_file,read_output_file
from utils import is_valid_network, average_pairwise_distance,average_pairwise_distance_fast
import os
import collections
import operator
import matplotlib.pyplot as plt
```

```python
# Run this cell to generate all outputs
if __name__ == "__main__":
    output_dir = "outputs"
    input_dir = "inputs"
    for input_path in os.listdir(input_dir):
        graph_name = input_path.split(".")[0]
        G = read_input_file(f"{input_dir}/{input_path}")
        T = SPTalgorithm(G)
    
        write_output_file(T, f"{output_dir}/{graph_name}.out")
```

```python
def SPTalgorithm(G):
    
    G = G.copy()
    T = nx.Graph()
   
    if len(G.nodes()) == 1:
        T.add_nodes_from(list(G.nodes())) 
        return T
    
    # find shortest path tree using dijskstra for each node
    # keep count of all edges being used in all shortest path trees
    # sort edges in decreasing order of appearance, 
    
    d = collections.defaultdict(int)
    for node in G.nodes():
        length, path = nx.single_source_dijkstra(G, node)
        if len(path) > 1:
            for p in path.values():
                for i in range(0, len(p) - 1):
                    edge = (p[i], p[i + 1])
                    d[edge] += 1    
    sorted_d = dict(sorted(d.items(), key=operator.itemgetter(1),reverse=True))  
    
    # add most appeared edges greedily
    # mark the nodes being covered by the edges (endpoints or neighbors)

    for edge in sorted_d.keys():
        p1 = edge[0]
        p2 = edge[1]
        f1 = False
        f2 = False
        if T.has_node(p1):
            f1 = True
        if T.has_node(p2):
            f2 = True
        if not f1:
            T.add_node(p1)
        if not f2:
            T.add_node(p2)
        T.add_edge(p1, p2)
        if len(nx.cycle_basis(T.to_undirected())) > 0:  # do not add the check if cause cycle
            if not f1:
                T.remove_node(p1)
            if not f2:
                T.remove_node(p2)
            T.remove_edge(p1, p2)
        else: 
            if is_valid_network(G, T):  # terminate when formed a valid network
                print(1)
                return T
```
